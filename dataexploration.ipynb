```python
#BILmap['FUNDING'] = BILmap['FUNDING'].astype('int')
BILmap = BILmap[BILmap['FUNDING'] != 'TBD']

countyData = BILmap.groupby(['COUNTY', 'STATE FIPS','STATE NAME'])['FUNDING'].sum().reset_index()
countyData
```

```python
# Assuming DataFrame A and DataFrame B are already defined 

# Iterate over each row in DataFrame A
for index, row in countyData.iterrows():
    county = row['COUNTY']
    stateFips = row['STATE FIPS']
    
    # Check if the corresponding city and state exist in DataFrame B
    matching_row = censusCounty[(censusCounty['NAME'] == county) & (censusCounty['STATEFP'] == stateFips)]
    
    # If a match is found, update the 'county' column in DataFrame A
    if len(matching_row) > 0:
        countyData.at[index, 'GEOID'] = matching_row.iloc[0]['GEOID']

# Print the updated DataFrame A
countyData
```
